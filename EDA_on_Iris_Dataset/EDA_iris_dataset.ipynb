# 2D- Scatterplot with color-coding for each flower type

```python
sns.set_style("whitegrid");
sns.FacetGrid(iris,hue="Species", height=4).map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend();
plt.show()
```

### Observations:

1. Using SepalLengthCm and SepalWidthCm features, we can distinguish setosa flowers from others.
2. Separating Versicolor from Viginica is much harder as they have considerable overlap.

# Pair - Plot

```python
plt.close()
sns.set_style("whitegrid");
sns.pairplot(iris,hue="Species", height = 3, diag_kind = "kde");
plt.show()
```

## OBSERVATIONS:

1. PetalLengthCm and PetalWidthCm are the most useful features to identify various flower types.
2. While Setosa can be easily identified (linearly separable), Vernica and Versicolor have some overlap(almost linearly separable).

# Mean, Variance and Standard Deviation

```python
print("Mean:")
print(np.mean(iris[iris['Species'] == 'Iris-setosa']['PetalLengthCm']))

print(np.mean(np.append(iris[iris['Species'] == 'Iris-setosa']['PetalLengthCm'],50)))
print(np.mean(iris[iris['Species'] == 'Iris-virginica']['PetalLengthCm']))
print(np.mean(iris[iris['Species'] == 'Iris-versicolor']['PetalLengthCm']))

print("\nVariance:")
print(np.var(iris[iris['Species'] == 'Iris-setosa']['PetalLengthCm']))
print(np.var(iris[iris['Species'] == 'Iris-virginica']['PetalLengthCm']))
print(np.var(iris[iris['Species'] == 'Iris-versicolor']['PetalLengthCm']))


print("\nStd-Dev:")
print(np.std(iris[iris['Species'] == 'Iris-setosa']['PetalLengthCm']))
print(np.std(iris[iris['Species'] == 'Iris-virginica']['PetalLengthCm']))
print(np.std(iris[iris['Species'] == 'Iris-versicolor']['PetalLengthCm']))
```

## OBSERVATIONS:

1. Scatterplot: Scatterplots are used to visualize the relationship between two continuous variables. They help us identify patterns such as correlation, clusters, or outliers in the data.

2. 
Boxplot: Boxplots are useful for visualizing the distribution of a continuous variable, showing its median, quartiles, and potential outliers. They provide a compact summary of the data's distribution and help in comparing distributions across different groups or categories

3. 
Violin plot: Violin plots are similar to boxplots but provide additional information about the probability density of the data at different values. They combine the features of a boxplot and a kernel density plot, giving a more detailed view of the data distributio

4.  .
Distplot (Distribution plot): Distplots display the distribution of a univariate set of observations, usually as a histogram. They also show an estimate of the probability density function (PDF) using either kernel density estimation (KDE) or a fitted parametric distributi

5.  n.
KDE (Kernel Density Estimation) plot: KDE plots provide a smooth estimate of the probability density function of a continuous random variable. They are particularly useful when visualizing the overall shape of the distribution, especially when the data is not well-represented by simple summary statistics.
